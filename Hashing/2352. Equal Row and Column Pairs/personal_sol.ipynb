## Problem
Given a 0-indexed n x n integer matrix grid, return the number of pairs (ri, cj) such that row ri and column cj are equal.

A row and column pair is considered equal if they contain the same elements in the same order (i.e., an equal array).

**Example 1:**
Input: grid = [[3,2,1],[1,7,6],[2,7,7]]  
Output: 1  
Explanation: There is 1 equal row and column pair:
- (Row 2, Column 1): [2,7,7]

**Example 2:**  
Input: grid = [[3,1,2,2],[1,4,4,5],[2,4,2,2],[2,4,2,2]]  
Output: 3  
Explanation: There are 3 equal row and column pairs:
- (Row 0, Column 0): [3,1,2,2]
- (Row 2, Column 2): [2,4,2,2]
- (Row 3, Column 2): [2,4,2,2]

**Constraints:**  
n == grid.length == grid[i].length  
1 <= n <= 200  
1 <= grid[i][j] <= 10<sup>5</sup>


```python
class Solution {
public:
    // create your own hash map!
    string convertToKey(vector<int>& arr) {
        string s = "";
        // for every number in array, create a unique key
        for (int num: arr) {
            s += to_string(num) + ",";
        }
        
        return s;
    }
    int equalPairs(vector<vector<int>>& grid) {
        unordered_map<string, int> rows;
        unordered_map<string, int> cols;
        int result = 0;

        for (auto row : grid) {
            // create unique row key
            string row_str = convertToKey(row);
            rows[row_str]++;
        }

        // There are total 3 columns (how many elements in row == how many columns)
        for (int i = 0; i < grid[0].size(); i++) {
            vector<int> col;
            // each element in grid == how many element in a single col
            for (int j = 0; j < grid.size(); j++) {
                // [first element, # of columns] in [every row] 
                col.push_back(grid[j][i]);
            }
            string col_str = convertToKey(col);
            cols[col_str]++;
        }

        for (auto pair : rows) {
            if (cols.contains(pair.first)) {
                result += pair.second * cols[pair.first];
            }
        }

        return result;
    }
};
```

```python
from typing import List
from collections import defaultdict

class Solution:
    def equalPairs(self, grid: List[List[int]]) -> int:
        def arr_to_tup(arr):
            return tuple(arr)
        rows = defaultdict(int)
        cols = defaultdict(int)

        for row in grid:
            rows[arr_to_tup(row)] += 1
        
        for col in range(len(grid[0])):
            col_arr = []
            for ele in range(len(grid)):
                col_arr.append(grid[ele][col])
            cols[arr_to_tup(col_arr)] += 1
        
        result = 0
        for arr in rows:
            result += rows[arr] * cols[arr]

        return result
        
```
